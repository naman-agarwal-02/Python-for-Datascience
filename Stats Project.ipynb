# 1

```python
df['Age1stCode'] = pd.to_numeric(df['Age1stCode'], errors='coerce')

aveage = df['Age1stCode'].mean()

print(f"Average age at which developers wrote their first line of code: {aveage:.2f} years")
```

# 2

```python
pythonuserscountry = df[df['LanguageWorkedWith'].str.contains('Python', na=False)]

percentage_users = (pythonuserscountry['Country'].value_counts() / df['Country'].value_counts()) * 100

print("Percentage of developers who know Python in each country:")
print(percentage_users)
```

# 3

```python
countriy_with_250 = df['Country'].value_counts()[df['Country'].value_counts() > 250].index

highest_hours_by_country = df[df['Country'].isin(countriy_with_250)] \
    .groupby('Country')['WorkWeekHrs'] \
    .mean() \
    .sort_values(ascending=False)

print("Countries where developers work the highest number of hours per week:")
print()
print(highest_hours_by_country)
```

```python
desired_languages_2020 = df['LanguageDesireNextYear'].str.split(';', expand=True).stack()

desired_languages_counts = desired_languages_2020.value_counts()

most_desired_language_2020 = desired_languages_counts.idxmax()

print("The most desired programming language for the year 2020:", most_desired_language_2020)
```

```python
hobbyist_distribution = df.groupby(['Country', 'Gender'])['Hobbyist'] \
    .value_counts(normalize=True) * 100

print("Distribution of people who code as a hobby based on gender and country:")

print(hobbyist_distribution)
```

```python
satisfaction_report = df.groupby(['Country', 'Gender', 'JobSat', 'CareerSat']).size() \
    .unstack(fill_value=0) \
    .apply(lambda x: x / x.sum() * 100, axis=1)

print("Job and career satisfaction of developers based on gender and country:")
print()
print()
print(satisfaction_report)
```
